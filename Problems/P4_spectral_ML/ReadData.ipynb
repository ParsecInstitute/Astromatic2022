```python
import matplotlib.pyplot as plt
import pickle
import mplcyberpunk
plt.style.use('cyberpunk')
```

## Read data
Here we will simply read in our pickle files that contain the spectra (inputs) and the fluxes (labels)

```python
# Spectra
spectra_SN3 = pickle.load(open('spectrum_SN3.pkl', 'rb'))
spectra_axis_SN3 = pickle.load(open('spectrum_axis_SN3.pkl', 'rb'))
spectra_SN2 = pickle.load(open('spectrum_SN2.pkl', 'rb'))
spectra_axis_SN2 = pickle.load(open('spectrum_axis_SN2.pkl', 'rb'))
spectra_SN1 = pickle.load(open('spectrum_SN1.pkl', 'rb'))
spectra_axis_SN1 = pickle.load(open('spectrum_axis_SN1.pkl', 'rb'))
# Labels
fluxes = pickle.load(open('fluxes.pkl', 'rb'))
```

```python
spec = 93

fig, (ax1, ax2, ax3) = plt.subplots(1, 3,  figsize=(25, 10), dpi=500)
fig.suptitle('SN3, SN2, and SN1 emission spectra for a respresentative image', weight='bold', fontsize=20)
ax1.plot(spectra_axis_SN3[spec], spectra_SN3[spec])
ax1.set_xlabel(r'wavenumber (cm$^{-1}$)', fontsize=18)
ax1.set_xlim(14700, 15400)
ax1.set_ylim(-0.1, 1.1)
ax2.plot(spectra_axis_SN2[spec], spectra_SN2[spec])
ax2.set_xlabel(r'wavenumber (cm$^{-1}$)', fontsize=18)
ax2.set_xlim(19000, 21000)
ax2.set_ylim(-0.02, 0.25)
ax3.plot(spectra_axis_SN1[spec], spectra_SN1[spec])
ax3.set_xlabel(r'wavenumber (cm$^{-1}$)', fontsize=18)
ax3.set_xlim(26000, 28000)
ax3.set_ylim(-0.01, 0.1)
mplcyberpunk.make_lines_glow()
```
